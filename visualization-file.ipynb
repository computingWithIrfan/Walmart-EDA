# Visualizing Cleaned data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data = pd.read_csv('./dataset/walmart_clean.csv')
data.head()
```

```python
data['date'] = pd.to_datetime(data['date'])
```

```python
plt.figure(figsize=(12, 6))
data.groupby(data['date'].dt.to_period('M'))['total'].sum().plot(kind='line', marker='o', color='b')
plt.title("Total Sales Over Time", fontsize=14,fontweight='bold')
plt.xlabel("Month", fontsize=12)
plt.ylabel("Total Sales", fontsize=12)
plt.grid(True)
plt.show()
```

---
---
---

```python
explode = [0.0, 0.08, 0.0, 0.0, 0.08, 0.0]
plt.figure(figsize=(6, 6))  # Increase figure size for better readability
data.groupby('category')['total'].sum().plot(
    kind='pie',
    autopct='%1.1f%%',       
    colormap='Pastel2',      
    shadow=True,            
    explode=explode,         
    startangle=90,          
    pctdistance=0.75,       
    labeldistance=1.05,
    textprops={'fontsize': 7} 
)
plt.title("Sales Distribution by Category", fontsize=14, fontweight="bold")
plt.show()
```

___
---
***

```python
plt.figure(figsize=(30, 10))
colors = sns.color_palette("crest", 100)  

data.groupby('branch')['total'].sum().sort_values(ascending=False).plot(kind='bar', color=colors)
plt.title('Total Sales per Branch', fontsize=28, fontweight='bold')
plt.xlabel('Branch', fontsize=16)
plt.ylabel('Total Sales', fontsize=16)
plt.yticks(np.arange(5000,26001,2000),fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

___
___
___

```python
plt.figure(figsize=(30, 10))
colors = sns.color_palette("inferno", 100)  

data.groupby('branch')['profit'].sum().sort_values(ascending=True).plot(kind='bar', color=colors)
plt.title('Total Profit generated per Branch', fontsize=28, fontweight='bold')
plt.xlabel('Branch', fontsize=16)
plt.ylabel('Total Profit', fontsize=16)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

___
___
___

```python
plt.figure(figsize=(5, 6))
data.groupby('category')[['profit','total']].sum().sort_values(by='total').plot(kind='barh',cmap='Set2')
plt.title('Profit and Sales by Category', fontsize=12, fontweight='bold')
plt.xlabel('Sales & Profit', fontsize=10)
plt.ylabel('Category', fontsize=10)
plt.xticks(np.arange(50000,500001,50000),fontsize=6)
plt.yticks(fontsize=10)
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

____
___
___

```python
plt.figure(figsize=(10, 6))
plt.hist(data['rating'], bins=20, color='skyblue', alpha=0.7)
plt.xticks(np.arange(3,10.1,0.5))
plt.yticks(np.arange(200,1801,200))
# Add labels and title with improved font style
plt.title('Distribution of Customer Ratings', fontsize=16, fontweight='bold')
plt.xlabel('Rating', fontsize=14)
plt.ylabel('Frequency', fontsize=14)

plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.tight_layout()

# Show the plot
plt.show()
```

___
___
___

```python
data.groupby('category')[['rating','quantity']].mean().sort_values(by='rating').plot(kind='barh',cmap='crest')
plt.title('Average Rating and Quanity Sold', fontsize=12, fontweight='bold')
plt.xlabel('Rating & Quantity ', fontsize=10)
plt.ylabel('Category', fontsize=10)
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

___
___
___

```python
data['hours'] = pd.to_datetime(data['time'], format='%H:%M:%S').dt.hour

# Group data by time (hour) and sum total sales
plt.figure(figsize=(10, 6))
data.groupby('hours')['total'].sum().plot(kind='line',marker='o', color='b', linestyle='-', alpha=0.7)
plt.title('Time vs Sales', fontsize=14)
plt.xlabel('Time (Hour)', fontsize=12)
plt.ylabel('Total Sales', fontsize=12)

plt.grid(True,alpha=0.2)
plt.xticks(range(24))  # Show hours from 0 to 23
plt.show()
```

___
___
___

```python
plt.figure(figsize=(18, 8))
colors = sns.color_palette("magma", 100)
data.groupby('city')['total'].sum().sort_values(ascending=False).plot(kind='bar', color = colors, edgecolor='black')
plt.title('City-wise Sales Distribution', fontsize=22, fontweight='bold')
plt.xlabel('City', fontsize=14)
plt.ylabel('Total Sales', fontsize=14)
plt.yticks(np.arange(5000,50001,5000),fontsize=12)
plt.xticks(fontsize=8)
plt.grid(axis='y',alpha=0.5)
plt.tight_layout()
plt.show()
```

___
___
___

```python
plt.figure(figsize=(9, 5))
data.groupby('category')['quantity'].sum().plot(kind='bar', color='lightgreen', edgecolor='black')

# Add labels and title
plt.title('Category-wise Quantity Sold', fontsize=13, fontweight='bold')
plt.xlabel('Category', fontsize=10)
plt.ylabel('Quantity Sold', fontsize=10)

# Display the plot
plt.xticks(rotation=45, ha='right', fontsize=12)  # Rotate x-axis labels for better readability
plt.yticks(np.arange(1500,10000,1000),fontsize=8)
plt.xticks(fontsize=8)
plt.grid(axis='y',alpha=0.5)
plt.show()
```
